```python
import os
import pandas as pd
```

```python
match_txt = open("Stats-2011.txt")
line_lst = match_txt.readlines()
```

```python
combined_txt = ''.join(line_lst)
```

```python
combined_txt.split('\n\n')[0]
```

```python
attribute_lst = ['MatchID', 'Possession_home',
 'Possession_away',
 'Shots_on_target_home',
 'Shots_on_target_away',
 'Shots_home',
 'Shots_away',
 'Touches_home',
 'Touches_away',
 'Passes_home',
 'Passes_away',
 'Tackles_home',
 'Tackles_away',
 'Clearances_home',
 'Clearances_away',
 'Corners_home',
 'Corners_away',
 'Offsides_home',
 'Offsides_away',
 'Yellow_cards_home',
 'Yellow_cards_away',
 'Fouls_conceded_home',
 'Fouls_conceded_away',
 'Red_cards_home',
 'Red_cards_away']

match_csv = open('Stats-2011.csv', 'w')

attribute_name = ','.join(attribute_lst) + '\n'

match_csv.write(attribute_name)

for row in combined_txt.split('\n\n'):
    row = row.replace("% ", "")
    data = row.split('\n')
    
    try:
        data.remove('')
    except:
        pass
    
    data_dict = {key : '' for key in attribute_lst}
    
    for i in range(len(data)):
        
        data_line = data[i].split(' ')
        if(i == 0):
            match_id = data_line[1]
            data_dict['MatchID'] = match_id
        else:
            home_data = data_line[0]
            away_data = data_line[-1]
            
            attribute = '_'.join(data_line[1:-1]) 
            
            attribute_h = attribute + "_home"
            attribute_a = attribute + "_away"
            
            data_dict[attribute_h] = home_data
            data_dict[attribute_a] = away_data
        
    new_line = ','.join(data_dict.values()) + '\n'
    match_csv.write(new_line)

match_csv.close()
        
```

```python
res_df = pd.read_csv('MatchResults-2011.csv')
stats_df = pd.read_csv('Stats-2011.csv')
```
