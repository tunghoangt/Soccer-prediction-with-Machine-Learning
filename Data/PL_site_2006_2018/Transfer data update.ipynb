```python
import pandas as pd
```

```python
masterdata = pd.read_csv('masterdata.csv')
transfer = pd.read_excel('Transfer information 2007-2018.xlsx')\

transfer = transfer[['Season', 'Club', 'Expenditures',
                   'Arrivals', 'Income', 'Departures']]

# Cleaning space

masterdata['Home_team'] = masterdata['Home_team'].str.strip()
masterdata['Away_team'] = masterdata['Away_team'].str.strip()

transfer['Club'] = transfer['Club'].str.strip()
```

```python
masterdata.head()
```

```python
transfer.head()
```

```python
merged_data = pd.merge(masterdata, transfer, how = 'left', 
                       left_on = ['year', 'Home_team'], right_on = ['Season', 'Club'])
```

```python
merged_data.head()
```

```python
merged_data.drop(['Season', 'Club'], axis = 1, inplace = True)
merged_data.rename(columns = {'Expenditures':'Expenditures_home', 'Arrivals':'Arrivals_home', 
                              'Income':'Income_home', 'Departures':'Departures_home'}, inplace = True)
```

```python
merged_data = pd.merge(merged_data, transfer, how = 'left', 
                       left_on = ['year', 'Away_team'], right_on = ['Season', 'Club'])
```

```python
merged_data.drop(['Season', 'Club'], axis = 1, inplace = True)
merged_data.rename(columns = {'Expenditures':'Expenditures_away', 'Arrivals':'Arrivals_away', 
                              'Income':'Income_away', 'Departures':'Departures_away'}, inplace = True)
```

```python
#Verify transfer no Null data
merged_data.isnull().values.any()
```

```python
merged_data.to_csv('masterdata1.csv', index = False)
```
